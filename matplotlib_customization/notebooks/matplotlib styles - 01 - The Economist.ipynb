## rcParams

rcParams are settings for matplotlib. They're how stylesheets work (the thing when you use `plt.style.use`.

You can find examples of the different styles [here](https://matplotlib.org/2.1.1/gallery/style_sheets/style_sheets_reference.html) and [here](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html).

> **A big long list is:** bmh, classic, dark_background, fivethirtyeight, ggplot, grayscale, seaborn-bright, seaborn-colorblind, seaborn-dark, seaborn-dark-palette, seaborn-darkgrid, seaborn-deep, seaborn-muted, seaborn-notebook, seaborn-paper, seaborn-pastel, seaborn-poster, seaborn-talk, seaborn-ticks, seaborn-white, seaborn-whitegrid 

If you think one of them is cool, you can find its rcparams styles by checking out its mplstyle file in [in this directory](https://github.com/matplotlib/matplotlib/tree/master/lib/matplotlib/mpl-data/stylelib).

# Line graph

```python
df = pd.read_csv("visas-by-continent-long.csv")
df = df.groupby(["year", "continent"]).visas.sum().unstack().reset_index()
df.drop(columns='Unknown', inplace=True)
df['year'] = pd.to_datetime(df.year, format="%Y")
df.set_index('year', inplace=True)
percent_df = df.div(df.sum(axis=1), axis='rows') * 100

percent_df.head(2)
```

# Scatter plot

```python
df = pd.read_csv("countries.csv")
df.head(2)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Don't touch this part of the code!
colors = list(reversed(plt.rcParams['axes.prop_cycle'].by_key()['color']))

groups = df.groupby('continent')
for name, group in groups:
    group.plot(kind='scatter', x='gdp_per_capita', y='life_expectancy', color=colors.pop(), label=name, s= 5 + (group.population / 200000), ax=ax, alpha=0.75)
ax.legend_.remove()

# CUSTOMIZATION CODE GOES UNDER HERE

plt.rcParams.update({
  'lines.linewidth': 4,
  'lines.solid_capstyle': 'butt',
  'font.family':'Andale Mono' ,
  'font.stretch':'extra-condensed',
  'legend.fancybox': False,
  'legend.frameon':False,
 
    'axes.prop_cycle': cycler('color', ['#02897e','#efb96a','#b6bd3b','#de6e6e','#f15b40','#00526d']),
  'axes.facecolor': '#FFFFFF',
  'axes.grid': True,
  'axes.labelsize': 'large',
  'axes.axisbelow': True,
  'ytick.major.pad': 10,
    'axes.labelpad': 20.0,
  'axes.edgecolor': 'black',
  'axes.linewidth': 1.0,
  'axes.xmargin':0.1,
    'axes.spines.left'   : False,
    'axes.spines.bottom' : False,
    'axes.spines.top'    : False,
    'axes.spines.right'  : False,
  
    'patch.edgecolor': '#f0f0f0',
  'patch.linewidth': 0.5,
  'svg.fonttype': 'path',
  'grid.linestyle': '-',
  'grid.linewidth': 1.0,
  'grid.color': '#cbcbcb',
    
    'ytick.left'           : True,
    'ytick.direction'      : 'out',
  'xtick.major.size': 0,
  'xtick.minor.size': 0,
  'ytick.major.size': 0,
  'ytick.minor.size': 0,
  'font.size': 14.0,
  'figure.facecolor': '#FFFFFF',
})

ax.set_xlabel("GDP per Capita")
ax.set_ylabel("Life Expectancy")
ax.grid(True)


# then we save it
plt.savefig("graph-scatter.pdf",bbox_inches='tight')
plt.show()
```
